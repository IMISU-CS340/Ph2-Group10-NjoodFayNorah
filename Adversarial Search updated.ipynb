# Main Concept and Objective:




The minimax algorithm is a way of finding an optimal move in a two player game. Alpha-beta pruning is a way of finding the optimal minimax solution while avoiding searching subtrees of moves which won't be selected. In the search tree for a two-player game, there are two kinds of nodes, nodes representing your moves and nodes representing your opponent's moves.

Alpha-Beta pruning is not actually a new algorithm, rather an optimization technique for minimax algorithm. It is a search algorithm that seeks to reduce the number of nodes that are evaluated by the minimum algorithm in the search tree. It produces the same output as minimax, but cuts off some branches that cannot influence the final decision - which greatly improves performance.

## Alpha Beta Algorithm :

### The main concept is to maintain two values through full research :
                                                                                                                                                          
•	Alpha: Best already explored option for player Max

•	Beta: Best already explored option for player Min

Initially, alpha is negative infinity and beta is positive infinity, i.e. in our code we'll be using the worst possible scores for both players.
Condition for Alpha-beta pruning:
The main condition which required for alpha-beta pruning is:
α>=β

We implement the first player as playing with maxAlpha (by calling minBeta).

## The evaluation function:   

We implement the evaluation function is a static number, that in accordance with the characteristics of the game itself, is being assigned to each node (position).

In the code below, we will be using an evaluation function that is fairly simple and common for all games in which it's possible to search the whole tree, all the way down to leaves.

It has 3 possible values:


	
- -1 if player that seeks minimum wins.
- 0 if it's a tie.
- 1 if player that seeks maximum wins.

```python
class FourWallsGame:
    
    def __init__(self):

#Start initializing the game.
# البدء بتهيئة اللعبة
        self.bulidingTheGame()
    
#------------------------------------------------------------------------------------------------------

#Method to build the game.
#.داله لبناء اللعبة
    def bulidingTheGame(self):
        
# build the board with 4*4 dimention array.
# بناء لوحة اللعب بحجم مصفوفه ٤*٤.
        self.board = [['.','.','.','.'],
                      ['.','.','.','.'],
                      ['.','.','.','.'],
                      ['.','.','.','.']]

# Red Player always plays first
# ًاللاعب الاحمر يبدء اللعب اولا 
        self.playerTurn = '🔴'
    
# Select the legle positions (starting from the bottem).
# تحديد المواقع المسموحه (بدايةً من القاع).
        self.A1=3
        self.A2=0
        
        self.B1=3
        self.B2=1
        
        self.C1=3
        self.C2=2
        
        self.D1=3
        self.D2=3 
#------------------------------------------------------------------------------------------------------

#Method to draw the board.
#دالة لرسم لوحة اللعب.
    def drawBoard(self):
        for i in range(0, 4):
            for j in range(0, 4):
                print('{}|'.format(self.board[i][j]), end=" ")
            print()
        print()
#------------------------------------------------------------------------------------------------------

#Determines if the made move is a legal move.
# دالة لتحديد الحركه اذا كانت مسموحه ام لا. 
    def checkLegalMove(self,i,j):
        
# check if the i and j out of range.
# التحقق من iو j .اذا كانت خارج الحدود
        if i < 0 or i > 3 or j < 0 or j > 3:
            return False

# check if the position is empty.
# التحقق بأن الموقع متاح.
        elif self.board[i][j] != '.':
            return False

# check if the position is legal.
# التحقق بأن الموقع مسموح.
        elif i == self.A1 and j == self.A2:
            return True
        elif i == self.B1 and j == self.B2:
            return True
        elif i == self.C1 and j == self.C2:
            return True
        elif i == self.D1 and j == self.D2:
            return True
        else:
            return False
        
#------------------------------------------------------------------------------------------------------
    
# method to return which letter for the column.
# دالة لارجاع الحرف الخاص بالعامود .
    def whichLegal(self,i,j):

        if i == self.A1 and j == self.A2:
            return 'A'
        elif i == self.B1 and j == self.B2:
            return 'B'
        elif i == self.C1 and j == self.C2:
            return 'C'
        elif i == self.D1 and j == self.D2:
            return 'D'
        
#------------------------------------------------------------------------------------------------------
    
# Checks if the game has ended and returns the winner if is there.
# .التحقق بأن اللعبة انتهت وارجاع اذا كان يوجد فائز
    def checkGameOver(self):
        
# check in vertical win.
# التحقق عند الفوز بشكل عامودي.
        for i in range(0, 4):
              if (self.board[0][i] != '.' and 
                  self.board[0][i] == self.board[1][i] and
                  self.board[1][i] == self.board[2][i] and
                  self.board[2][i] == self.board[3][i]):
                    return self.board[0][i] 
                
# check in horizontal win.
# التحقق عند الفوز بشكل افقي.
        for i in range(0, 4): 
            
            if (self.board[i] == ['🔴', '🔴', '🔴', '🔴']):
                return '🔴'
 
            elif (self.board[i] == ['🔵', '🔵', '🔵','🔵']):
                return '🔵'
        
# check with diagonal win in the left.
# التحقق عند الفوز بشكل قطري من جهة اليسار
        if (self.board[0][0] != '.' and 
            self.board[0][0] == self.board[1][1] and 
            self.board[0][0] == self.board[2][2] and 
            self.board[0][0] == self.board[3][3]):
            return self.board[0][0]
        
# check with diagonal win in the right.
# التحقق عند الفوز بشكل قطري من جهة اليمين.
        if (self.board[0][3] != '.' and 
            self.board[0][3] == self.board[1][2] and 
            self.board[0][3] == self.board[2][1] and
            self.board[0][3] == self.board[3][0] ):
            return self.board[0][3]
        
# check if the whole board full.
# التحقق اذا اللوحة ممتلئه.
        for i in range(0, 4):
            for j in range(0, 4):
                
# There's an empty field, we continue the game.
# التحقق ما اذا كان هناك مكان فارغ لاستمرار اللعب
                if (self.board[i][j] == '.'):
                    return None
        
# if is no winner then return an empty.
# اذا لم يكن هناك فائز الداله ترجع فراغ.
        return '.'
  
#------------------------------------------------------------------------------------------------------
# 🔵
# alpha method
# دالة الفا
    def maxAlpha(self, alpha, beta):
        
# initialize maxv setting it to -2 as worse than the worst case.
# القيمه الابتدائيه للقيمة القصوى في اسوأ الحالات .
        maxv = -2  # -∞
        pi = None
        pj = None
        
# check if the game is over.
#التحقق اذا اللعبة انتهت .
        result = self.checkGameOver()

# If the game came to an end, the function needs to return the evaluation function for the blue player 
#of the end and that can be: 
# -1  loss, 0  a no winner and 1 win.
#  اذا اللعبة انتهت فيتم ارجاع دالة التقييم بالنسبة للاعب الازرق وتكون:
# ١ فائز ،٠ تعادل و -١ خاسر
        if result == '🔴':
            return (-1, 0, 0)
        elif result == '🔵':
            return (1, 0, 0)
        elif result == '.':
            return (0, 0, 0)

# insted loop for evalute the alpha beta in each case.
# حلقات مركبه لحساب الفا بيتا لكل موقع محتمل.
        for i in range(0, 4):
            for j in range(0, 4):
                if self.board[i][j] == '.' and self.checkLegalMove(i,j):
                    
# On the empty field Blue player makes a move and calls beta (one branch of the game tree).
#  لكل موقع متاح اللاعب الازرق يلعب حركته (يعتبر فرع واحد من الشجرة) ويقوم بمناداة دالة بيتا
                    self.board[i][j] = '🔵'

# to return which letter for the column and update the value.
#  لارجاع الحرف الخاص بالعامود لتحديث القيم 
                    s = self.whichLegal(i,j)
                    if s == 'A': 
                        self.A1 = i-1
                        self.A2 = j
                    elif s == 'B': 
                        self.B1 = i-1
                        self.B2 = j
                    elif s == 'C': 
                        self.C1 = i-1
                        self.C2 = j
                    else: 
                        self.D1 = i-1
                        self.D2 = j
                    
                    (Ev, min_i, min_j) = self.minBeta(alpha, beta)
                    
# updating the maxv value if needed.
# تحديث القيمة الكبرى اذا تحقق الشرط
                    if Ev > maxv:
                        maxv = Ev
                        pi = i
                        pj = j

# Setting back the field to empty and the valus.
#  نرجع الحركه والقيم كما كانت .
                    self.board[i][j] = '.'
                    if s == 'A':
                        self.A1=i
                        self.A2=j
                    elif s == 'B':
                        self.B1=i
                        self.B2=j
                    elif s == 'C': 
                        self.C1=i
                        self.C2=j
                    else: 
                        self.D1=i
                        self.D2=j  
                    
# check if the maximum value is greater or equal than beta then cut the tree (cut beta).
# تحقق مما إذا كانت القيمة الكبرى أكبر أو تساوي من بيتا ثم يتم قطع الفرع من الشجرة (قطع بيتا).
                    if maxv >= beta:
                        return (maxv, pi, pj)

# check if the maximum value is greater than alpha then update the value of alpha.
# التحقق مما إذا كانت القيمة الكبرى أكبر من ألفا فيتم تحديث قيمة ألفا
                    if maxv > alpha:
                        alpha = maxv
       
        return (maxv, pi, pj)
    
#------------------------------------------------------------------------------------------------------
# 🔴
# beta method
# دالة الفا
    def minBeta(self, alpha, beta):
        
# initialize the minv setting it to 2 as worse than the worst case.
# القيمه الابتدائيه للقيمة الصغرى في اسوأ الحالات .
        minv = 2  # ∞
        qi = None
        qj = None

# check if the game is over.
#التحقق اذا اللعبة انتهت .
        result = self.checkGameOver()

# If the game came to an end, the function needs to return the evaluation function for the red player  
#of the end and that can be: 
# -1  win, 0  a no winner and 1 loss.
#  اذا اللعبة انتهت فيتم ارجاع دالة التقييم بالنسبة للاعب الاحمر وتكون:
# ١خاسر ، ٠ تعادل و -١ فائز
        if result == '🔴':
            return (-1, 0, 0)
        elif result == '🔵':
            return (1, 0, 0)
        elif result == '.':
            return (0, 0, 0)

# insted loop for evalute the alpha beta in each case.
# حلقات مركبه لحساب الفا بيتا لكل موقع محتمل.
        for i in range(0, 4):
            for j in range(0, 4):
                if self.board[i][j] == '.' and self.checkLegalMove(i,j):
                   
 #On the empty field red player makes a move and calls alpha (one branch of the game tree).
#  لكل موقع متاح اللاعب الاحمر يلعب حركته (يعتبر فرع واحد من الشجرة) ويقوم بمناداة دالة الفا
                    self.board[i][j] = '🔴'
                    
# to return which letter for the column and update the value.
#  لارجاع الحرف الخاص بالعامود لتحديث القيم
                    s = self.whichLegal(i,j)
                    if s == 'A': 
                        self.A1 = i-1
                        self.A2 = j
                    elif s == 'B': 
                        self.B1 = i-1
                        self.B2 = j
                    elif s == 'C': 
                        self.C1 = i-1
                        self.C2 = j
                    else: 
                        self.D1 = i-1
                        self.D2 = j
                                
                    (Ev, max_i, max_j) = self.maxAlpha(alpha, beta)
                                    
# updating the minv value if needed.
# تحديث القيمة الادنى اذا تحقق الشرط
                    if Ev < minv:
                        minv = Ev
                        qi = i
                        qj = j

# Setting back the field to empty and the valus.
#  نرجع الحركه والقيم كما كانت اذا انتهى حساب الفا
                    self.board[i][j] = '.'
                    if s == 'A':
                        self.A1=i
                        self.A2=j
                    elif s == 'B':
                        self.B1=i
                        self.B2=j
                    elif s == 'C': 
                        self.C1=i
                        self.C2=j
                    else:
                        self.D1=i
                        self.D2=j


# check if the minimum value is less or equal than alpha then cut the tree (cut alpha).
# تحقق مما إذا كانت القيمة الادنى اقل أو تساوي من الفا ثم يتم قطع الفرع من الشجرة (قطع الفا).
                    if minv <= alpha:
                        return (minv, qi, qj)

# check if the minimum value is less than beta then update the value of beta.
# التحقق مما إذا كانت القيمة الادنى اصغر من بيتا فيتم تحديث قيمة بيتا
                    if minv < beta:
                        beta = minv
       
        return (minv, qi, qj)
    
#------------------------------------------------------------------------------------------------------

# method for playing the game.
# داله للعب اللعبة
    def playAlphaBeta(self):
      
 # looping for exchange the players and continue the game.
 # حلقات لتبادل اللاعبين واستمرار اللعبة.
        while True:

# calling the drawBoard and checkGameOver methods.
#  مناداة دوال رسم اللعبة و التحقق من اللعب. 
            self.drawBoard()
            self.result = self.checkGameOver()

# check if there is a winner or tied to end the game.
# تحقق مما إذا كان هناك فائز او تعادل لإنهاء المباراة.
            if self.result != None:
 
 # print the winner and its evalution function.
 # .طباعة الفائز و دالة التقييم له
                if self.result == '🔴':
                    print('The winner is 🔴!')
                    print('The evalution function is:{}'.format(Ev))
                    return
                elif self.result == '🔵':
                    print('The winner is 🔵!')
                    print('The evalution function is:{}'.format(Ev))
                    return
                elif self.result == '.':
                    print("No winner!")
                    print('The evalution function is:{}'.format(Ev))
                    return

# check if the player is red.
# التحقق اذا كان اللاعب هو الاحمر
            if self.playerTurn == '🔴':

                while True:
                
 # calculate alpha and beta to give hent for the user.
 #حساب ألفا وبيتا لاعطاء تلميح للمستخدم.    
                    (Ev, qi, qj) = self.minBeta(-2, 2)
                    print('The recommended move after calculating α and β: Row = {}, Column = {}'.format(qi, qj))

# take the input from the user.
# اخذ المدخلات من المستخدم.
                    i = int(input('Insert the raw: '))
                    j = int(input('Insert the column: '))

# check if the input is legal move.
# التحقق اذا كانت الحركه مسموحه.
                    if self.checkLegalMove(i, j):
             
# to return which letter for the column and update the value.
#  لارجاع الحرف الخاص بالعامود لتحديث القيم.
                        s = self.whichLegal(i,j)
                        if s == 'A': 
                            self.A1 = i-1
                            self.A2 = j
                        elif s == 'B': 
                            self.B1 = i-1
                            self.B2 = j
                        elif s == 'C': 
                            self.C1 = i-1
                            self.C2 = j
                        else:
                            self.D1 = i-1
                            self.D2 = j
                            
# play the move.
# اللعب بوضع الحركه.
                        print("Your turn:")
                        self.board[i][j] = '🔴'
    
# exchange the player turn.
# تبديل دور اللاعب 
                        self.playerTurn = '🔵'
                        break
                    else:
                        print('The move is invalid! Try again.')

# else if is blue player turn.
#  اما اذا كان دور اللاعب الازرق
            else:
                (Ev, pi, pj) = self.maxAlpha(-2, 2)
    
# to return which letter for the column and update the value.
#  لارجاع الحرف الخاص بالعامود لتحديث القيم
                s = self.whichLegal(pi,pj)
                if s == 'A': 
                    self.A1 = pi-1
                    self.A2 = pj
                elif s == 'B': 
                    self.B1 = pi-1
                    self.B2 = pj
                elif s == 'C': 
                    self.C1 = pi-1
                    self.C2 = pj
                else:
                    self.D1 = pi-1
                    self.D2 = pj
  
 # play the move.
# اللعب بوضع الحركه.
                print("AI turn:")
                self.board[pi][pj] = '🔵'

# exchange the player turn.
# تبديل دور اللاعب 
                self.playerTurn = '🔴'
```

```python
class TicTocTieGame :
    
    def __init__(self):

#Start initializing the game.
# البدء بتهيئة اللعبة
        self.bulidingTheGame()
    
#------------------------------------------------------------------------------------------------------

#Method to build the game.
#.داله لبناء اللعبة
    def bulidingTheGame(self):
        
# build the board with 3*3 dimention array.
# بناء لوحة اللعب بحجم مصفوفه 3*3.
        self.board = [['.','.','.'],
                      ['.','.','.'],
                      ['.','.','.']]
                

#❌ always plays first
# ً  يبدء اللعب اولا❌ 
        self.playerTurn = '❌'
    
#------------------------------------------------------------------------------------------------------

#Method to draw the board.
#دالة لرسم لوحة اللعب.
    def drawBoard(self):
        for i in range(0, 3):
            for j in range(0, 3):
                print('{} |'.format(self.board[i][j]), end=" ")
            print()
        print()
#------------------------------------------------------------------------------------------------------

#Determines if the made move is a legal move.
# دالة لتحديد الحركه اذا كانت مسموحه ام لا. 
    def checkLegalMove(self,i,j):
        
# check if the i and j out of range.
# التحقق من iو j .اذا كانت خارج الحدود
        if i < 0 or i > 2 or j < 0 or j > 2:
            return False

# check if the position is empty.
# التحقق بأن الموقع متاح.
        elif self.board[i][j] != '.':
            return False

        else:
            return True
           
#------------------------------------------------------------------------------------------------------
    
# Checks if the game has ended and returns the winner if is there.
# .التحقق بأن اللعبة انتهت وارجاع اذا كان يوجد فائز
    def checkGameOver(self):
        
# check in vertical win.
# التحقق عند الفوز بشكل عامودي.
        for i in range(0, 3):
              if (self.board[0][i] != '.' and 
                  self.board[0][i] == self.board[1][i] and
                  self.board[1][i] == self.board[2][i]): 
                    return self.board[0][i] 
                
# check in horizontal win.
# التحقق عند الفوز بشكل افقي.
        for i in range(0, 3): 
            
            if (self.board[i] == ['❌', '❌', '❌']):
                return '❌'
 
            elif (self.board[i] == ['⭕️', '⭕️', '⭕️']):
                return '⭕️'
        
# check with diagonal win in the left.
# التحقق عند الفوز بشكل قطري من جهة اليسار
        if (self.board[0][0] != '.' and 
            self.board[0][0] == self.board[1][1] and 
            self.board[0][0] == self.board[2][2] ):
            return self.board[0][0]
        
# check with diagonal win in the right.
# التحقق عند الفوز بشكل قطري من جهة اليمين.
        if (self.board[0][2] != '.' and 
            self.board[0][2] == self.board[1][1] and 
            self.board[0][2] == self.board[2][0] ):
            return self.board[0][2]
        
# check if the whole board full.
# التحقق اذا اللوحة ممتلئه.
        for i in range(0, 3):
            for j in range(0, 3):
                
# There's an empty field, we continue the game.
# التحقق ما اذا كان هناك مكان فارغ لاستمرار اللعب
                if (self.board[i][j] == '.'):
                    return None
        
# if is no winner then return an empty.
# اذا لم يكن هناك فائز الداله ترجع فراغ.
        return '.'
  
#------------------------------------------------------------------------------------------------------
# ⭕️
# alpha method
# دالة الفا
    def maxAlpha(self, alpha, beta):
        
# initialize maxv setting it to -2 as worse than the worst case.
# القيمه الابتدائيه للقيمة القصوى في اسوأ الحالات .
        maxv = -2  # -∞
        pi = None
        pj = None
        
# check if the game is over.
#التحقق اذا اللعبة انتهت .
        result = self.checkGameOver()

# If the game came to an end, the function needs to return the evaluation function for the ⭕️
#of the end and that can be: 
# -1  loss, 0  a no winner and 1 win.
#  اذا اللعبة انتهت فيتم ارجاع دالة التقييم بالنسبة ⭕️ وتكون:
# ١ فائز ،٠ تعادل و -١ خاسر
        if result == '❌':
            return (-1, 0, 0)
        elif result == '⭕️':
            return (1, 0, 0)
        elif result == '.':
            return (0, 0, 0)

# insted loop for evalute the alpha beta in each case.
# حلقات مركبه لحساب الفا بيتا لكل موقع محتمل.
        for i in range(0, 3):
            for j in range(0, 3):
                if self.board[i][j] == '.' :
                    
# On the empty field ⭕️ makes a move and calls beta (one branch of the game tree).
#  لكل موقع متاح ⭕️ يلعب حركته (يعتبر فرع واحد من الشجرة) ويقوم بمناداة دالة بيتا
                    self.board[i][j] = '⭕️'
                    
                    (Ev, min_i, min_j) = self.minBeta(alpha, beta)
                    
# updating the maxv value if needed.
# تحديث القيمة الكبرى اذا تحقق الشرط
                    if Ev > maxv:
                        maxv = Ev
                        pi = i
                        pj = j

# Setting back the field to empty .
#  نرجع الحركه كما كانت .
                    self.board[i][j] = '.'
                    
# check if the maximum value is greater or equal than beta then cut the tree (cut beta).
# تحقق مما إذا كانت القيمة الكبرى أكبر أو تساوي من بيتا ثم يتم قطع الفرع من الشجرة (قطع بيتا).
                    if maxv >= beta:
                        return (maxv, pi, pj)

# check if the maximum value is greater than alpha then update the value of alpha.
# التحقق مما إذا كانت القيمة الكبرى أكبر من ألفا فيتم تحديث قيمة ألفا
                    if maxv > alpha:
                        alpha = maxv
       
        return (maxv, pi, pj)
    
#------------------------------------------------------------------------------------------------------
#❌
# beta method
# دالة الفا
    def minBeta(self, alpha, beta):
        
# initialize the minv setting it to 2 as worse than the worst case.
# القيمه الابتدائيه للقيمة الصغرى في اسوأ الحالات .
        minv = 2  # ∞
        qi = None
        qj = None

# check if the game is over.
#التحقق اذا اللعبة انتهت .
        result = self.checkGameOver()

# If the game came to an end, the function needs to return the evaluation function for the ❌ 
#of the end and that can be: 
# -1  win, 0  a no winner and 1 loss.
#  اذا اللعبة انتهت فيتم ارجاع دالة التقييم بالنسبة❌ وتكون:
# ١خاسر ، ٠ تعادل و -١ فائز
        if result == '❌':
            return (-1, 0, 0)
        elif result == '⭕️':
            return (1, 0, 0)
        elif result == '.':
            return (0, 0, 0)

# insted loop for evalute the alpha beta in each case.
# حلقات مركبه لحساب الفا بيتا لكل موقع محتمل.
        for i in range(0, 3):
            for j in range(0, 3):
                if self.board[i][j] == '.':
                   
 #On the empty field ❌ makes a move and calls alpha (one branch of the game tree).
#  لكل موقع متاح ❌ يلعب حركته (يعتبر فرع واحد من الشجرة) ويقوم بمناداة دالة الفا
                    self.board[i][j] = '❌'
                                
                    (Ev, max_i, max_j) = self.maxAlpha(alpha, beta)
                                    
# updating the minv value if needed.
# تحديث القيمة الادنى اذا تحقق الشرط
                    if Ev < minv:
                        minv = Ev
                        qi = i
                        qj = j

# Setting back the field to empty .
#  نرجع الحركه كما كانت اذا انتهى حساب الفا
                    self.board[i][j] = '.'

# check if the minimum value is less or equal than alpha then cut the tree (cut alpha).
# تحقق مما إذا كانت القيمة الادنى اقل أو تساوي من الفا ثم يتم قطع الفرع من الشجرة (قطع الفا).
                    if minv <= alpha:
                        return (minv, qi, qj)

# check if the minimum value is less than beta then update the value of beta.
# التحقق مما إذا كانت القيمة الادنى اصغر من بيتا فيتم تحديث قيمة بيتا
                    if minv < beta:
                        beta = minv
       
        return (minv, qi, qj)
    
#------------------------------------------------------------------------------------------------------

# method for playing the game.
# داله للعب اللعبة
    def playAlphaBeta(self):
      
 # looping for exchange the players and continue the game.
 # حلقات لتبادل اللاعبين واستمرار اللعبة.
        while True:

# calling the drawBoard and checkGameOver methods.
#  مناداة دوال رسم اللعبة و التحقق من اللعب. 
            self.drawBoard()
            self.result = self.checkGameOver()

# check if there is a winner or tied to end the game.
# تحقق مما إذا كان هناك فائز او تعادل لإنهاء المباراة.
            if self.result != None:
 
 # print the winner and its evalution function.
 # .طباعة الفائز و دالة التقييم له
                if self.result == '❌':
                    print('The winner is ❌!')
                    print('The evalution function is:{}'.format(Ev))
                    return
                elif self.result == '⭕️':
                    print('The winner is ⭕️!')
                    print('The evalution function is:{}'.format(Ev))
                    return
                elif self.result == '.':
                    print("No winner!")
                    print('The evalution function is:{}'.format(Ev))
                    return

# check if the player is ❌.
# التحقق اذا كان اللاعب هو ❌
            if self.playerTurn == '❌':

                while True:
                
 # calculate alpha and beta to give hent for the user.
 #حساب ألفا وبيتا لاعطاء تلميح للمستخدم.    
                    (Ev, qi, qj) = self.minBeta(-2, 2)
                    print('The recommended move after calculating α and β: Row = {}, Column = {}'.format(qi, qj))

# take the input from the user.
# اخذ المدخلات من المستخدم.
                    print("Your turn:")
                    i = int(input('Insert the raw: '))
                    j = int(input('Insert the column: '))

# check if the input is legal move.
# التحقق اذا كانت الحركه مسموحه.
                    if self.checkLegalMove(i, j):
                            
# play the move.
# اللعب بوضع الحركه.         
                        self.board[i][j] = '❌'
    
# exchange the player turn.
# تبديل دور اللاعب 
                        self.playerTurn = '⭕️'
                        break
                    else:
                        print('The move is invalid! Try again.')

# else if is ⭕️ turn.
#  اما اذا كان دور ⭕️
            else:
                (Ev, pi, pj) = self.maxAlpha(-2, 2)
 # play the move.
# اللعب بوضع الحركه.
                print("AI turn:")
                self.board[pi][pj] = '⭕️'

# exchange the player turn.
# تبديل دور اللاعب 
                
                self.playerTurn = '❌'
```

ــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــ
